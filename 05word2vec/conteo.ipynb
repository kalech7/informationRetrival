```python
from collections import Counter
from nltk.util import ngrams

# Función para generar n-gramas
def generar_n_gramas(palabras, n):
    # Crear n-gramas a partir de la lista de palabras
    n_gramas = list(ngrams(palabras, n))
    # Contar la frecuencia de cada n-grama
    return Counter(n_gramas)

# Definir el valor de n para los n-gramas
n = 2  # Cambia a 3 para trigramas, 4 para 4-gramas, etc.

# Generar n-gramas de las palabras después del stemming
n_gramas_frecuencia = generar_n_gramas(palabras_stem, n)

# Mostrar los n-gramas más comunes
top_n = 50# Número de n-gramas a mostrar
print(f"Los {top_n} {n}-gramas más comunes son:")
for n_grama, frecuencia in n_gramas_frecuencia.most_common(top_n):
    print(f"{n_grama}: {frecuencia}")
```
