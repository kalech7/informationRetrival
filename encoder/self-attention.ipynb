```python
from sklearn.feature_extraction.text import CountVectorizer
corpus= ['El perro juega con la pelota', 'el gato esta en la casa ', 'la pelota rueda sobre el piso de la casa', 'el gato juega con la pelota','el perro juega con el gato']
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus).toarray()
# Diccionario de términos
vocab = vectorizer.get_feature_names_out()
print("Vocabulario:", vocab)
```

```python
import numpy as np

# Dimensión de los vectores
n, d = X.shape

# Matrices de proyección
Wq = np.random.rand(d, d)  # Matriz de proyección para Q
Wk = np.random.rand(d, d)  # Matriz de proyección para K
Wv = np.random.rand(d, d)  # Matriz de proyección para V

print("Matriz de proyección para Q:\n", Wq)

# Cálculo de Q, K y V usando np.dot
Q = np.dot(X, Wq)
K = np.dot(X, Wk)
V = np.dot(X, Wv)

# Similaridad entre las frases usando np.dot
scores = np.dot(Q, K.T)  # Producto escalar entre Q y K^T

# Cálculo de pesos de atención usando la fórmula proporcionada
attention_weights = np.exp(scores) / np.sum(np.exp(scores), axis=-1, keepdims=True)

# Representación final después de self-attention
Z = np.dot(attention_weights, V)

print("Pesos de atención:\n", attention_weights)
print("Representación final (Z):\n", Z)
```
